### doing one by one is soo time consuming, so lets automate the process

```python
columns_with_nan = df.isnull().any()
df.columns[columns_with_nan] # returned true values
```

```python
# Imputing Numerical features with Mean
numerical_features = df.select_dtypes(include='float64')
numerical_features = numerical_features.columns[numerical_features.isnull().any()]
numerical_features # nan numerical features
```

```python
df[numerical_features] = df[numerical_features].fillna(df[numerical_features].mean())
```

```python
# Imputing Categorial features with Mode
categorical_features = df.select_dtypes(include='object')
categorical_features = categorical_features.columns[categorical_features.isnull().any()]
categorical_features
```

```python
df[categorical_features] = df[categorical_features].fillna(df[categorical_features].mode().iloc[0])
```

```python
df.isnull().sum()
```

```python
df[df.index == 133]
```

```python
df[categorical_features].mode()
```

## Remove unnecessary special characters

```python
df['Capital/Major City'] = df['Capital/Major City'].str.replace(r'[^a-zA-Z0-9\s]',"",regex=True)
df['Currency-Code'] = df['Currency-Code'].str.replace(r'[^a-zA-Z0-9\s]',"",regex=True)
df['Largest city'] = df['Largest city'].str.replace(r'[^a-zA-Z0-9\s]',"",regex=True)
df['Official language'] = df['Official language'].str.replace(r'[^a-zA-Z0-9\s]',"",regex=True)
```

## Top 20 Highest Populated Countries

```python
pop_sorted_df  = df.sort_values(by='Population', ascending=False)
top20_populated_countries = pop_sorted_df.head(20)
```

```python
# horizontal bar plot
pop_ax = sns.barplot(data=top20_populated_countries, x='Population',y='Country',hue='Country',palette='gist_rainbow_r',saturation=0.7)
for p in pop_ax.patches:
    pop_ax.annotate(f'{p.get_width():,.0f}', (p.get_width(), p.get_y() + p.get_height() / 2),
                     xytext=(5, 0), textcoords='offset points', ha='left', va='center', fontsize=10)
```

> As per this 2023 report, India and China are the most populated countries in the world followed by US

## Top 20 Most Densed Countries

```python
dense_sorted_df = df.sort_values(by='P_density', ascending=False)
top20_densed_countries = dense_sorted_df.head(20)
```

```python
dense_ax = sns.barplot(data=top20_densed_countries, x='P_density', y='Country', hue='Country',palette='Spectral')
dense_ax.set_title("Top 20 Most Densed Countries")
dense_ax.set_xticklabels(dense_ax.get_xticklabels(), rotation = 45,ha='right')
for p in dense_ax.patches:
    dense_ax.annotate(f'{p.get_width():,.0f}', (p.get_width(), p.get_y() + p.get_height() / 2),
                     xytext=(5, 0), textcoords='offset points', ha='left', va='center', fontsize=10)
plt.show()
```

> Monaco is the most densely populated country in the world follwed by Singapore.
